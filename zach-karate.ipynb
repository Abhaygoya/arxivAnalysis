```python
#Import networkx and some functions defined for network analysis and visualization
import networkx as nx
from functions import *
from shapleyFunctions import *
```

```python
G, pos = createZachary()

#Shift all node labels by 1
G = nx.relabel_nodes(G, lambda x: x+1)
pos = nx.spring_layout(G)

#Draw the network
# draw(G, pos, nx.betweenness_centrality(G), "Betweenness Centrality")
nx.draw(G, pos, with_labels=True)
```

```python
#Degree centrality
draw(G, pos, nx.degree_centrality(G), "Degree Centrality")
```

```python
#Closeness centrality
draw(G, pos, nx.closeness_centrality(G), "Closeness Centrality")
```

```python
#Betweenness centrality
draw(G, pos, nx.betweenness_centrality(G), "Betweenness Centrality")
```

```python
#Eigenvector centrality
draw(G, pos, nx.eigenvector_centrality(G), "Eigenvector Centrality")
```

```python
#Current flow centrality
draw(G, pos, nx.current_flow_closeness_centrality(G), "Current Flow Closeness Centrality")
```

```python
#Current flow betweenness centrality
draw(G, pos, nx.current_flow_betweenness_centrality(G), "Current Flow Betweenness Centrality")
```

```python
#Shapley Game 1
draw(G, pos, shapleyGame1(G), "Shapley Game 1")
```

```python
#Shapley Game 2
draw(G, pos, shapleyGame2(G, 2), "Shapley Game 2")
```

```python
#Shapeley Betweenness
draw(G, pos, ShapelyBetweeness(G), "Shapley Betweenness")
```
